# Importação de bibliotecas utilizadas no projeto

```python
import graphlab
import matplotlib.pyplot as plt
```

# Mapeamento dos arquivos de dados e carregamento

```python
path = '/Users/Marku/Documents/data_center/Workspace/DATA/movie_ratings/ml-20m/'

movies = graphlab.SFrame( path + 'movies.csv' )
ratings = graphlab.SFrame( path + 'ratings.csv' )
```

# Data Cleaning dos dados

```python
split_movie_year  = lambda title : title.split('(')[1].split(')')[0]
split_movie_title = lambda title : title.split('(')[0].split(')')[0].strip()
```

```python
movies['year']  = movies['title'].apply( split_movie_year )
movies['title'] = movies['title'].apply( split_movie_title )

movies.head()
```

# Sumariza e agrupa por MovieId o dataset de avaliações

```python
ratings_grouped = ratings.groupby(['movieId'], {'count':graphlab.aggregate.COUNT('userId'),
                                                'mean' :graphlab.aggregate.MEAN('rating'),
                                                'max': graphlab.aggregate.MAX('rating') ,
                                                'min': graphlab.aggregate.MIN('rating'),
                                                'stdv' : graphlab.aggregate.STDV('rating') }  )
```

# Data Exploration (Plota a dispersão das features count e mean)

```python
graphlab.canvas.set_target('ipynb')
ratings_grouped.show(view='Scatter Plot',x='count',y='mean')
```

# Elaboração do modelo de regressão linear

## Separação dos datasets de treinamento e teste

```python
train_data,test_data = ratings_grouped.random_split(fraction=.8,seed=0)
```

# Cria dois modelos regressão para a variável dependente mean (média de avaliações)

## @features_model: possui como variáveis dependentes as features [count,stdv,max,min]
## @count_model: possui como variávei dependente apenas a feature [count]

```python
feat = ['count','stdv','max','min']
features_model = graphlab.linear_regression.create( train_data , target = 'mean',features=feat,validation_set=None)
count_model    = graphlab.linear_regression.create( train_data , target = 'mean',features=['count'] ,validation_set=None)
```

# Plota os dois modelos lineares

```python
%matplotlib inline
plt.Figure()
plt.plot(test_data['count'],test_data['mean'],'.',
         test_data['count'],features_model.predict(test_data),'o',
         test_data['count'],count_model.predict(test_data),'x')
```

# Avaliação dos modelos

```python
print features_model.evaluate(test_data)
print count_model.evaluate(test_data)

test_data.topk('count',20)
```
