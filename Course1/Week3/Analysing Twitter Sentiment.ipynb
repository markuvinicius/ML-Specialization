# Cleaning data

```python
#cleans the "b carachter at the beginning and " at the end of tweet text

twitter_data['text'] = twitter_data['text'].apply(lambda t : t[1::][1:-2])

#clean the b carachter at the beginning of the username column
twitter_data['username'] = twitter_data['username'].apply(lambda t : t.split("'")[1])
```

# Build the Word Count Vector

```python
twitter_data['word_count'] = graphlab.text_analytics.count_words( sa=twitter_data['text'] , to_lower=True )
```
